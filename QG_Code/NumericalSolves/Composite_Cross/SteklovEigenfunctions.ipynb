# Steklov Eigenfunctions

Script will attempt to determine Steklov eigenvalues and eigenfunctions that correspond to seperable solutions.

Let $\Omega=\left[0,\frac{1}{2}\right)^2$, $\omega^2>0$, and $\theta\in\left[-\pi,\pi\right)^2$.
Note the mismatch between $\Omega$ and $\theta$ since this is one of the "subregions" of our Composite-Cross example.
The Steklov eigenvalue problem in $\Omega$ is to find $\lambda\in\mathbb{C}$ such that
\begin{align}
    \Delta_{\theta}u + \omega^2 u &= 0, &\qquad\text{in } \Omega, \\
    u\vert_{\partial\Omega} &= \lambda\dfrac{\partial u}{\partial n}\big\vert_{\partial\Omega}, &\qquad\text{on } \partial\Omega.
\end{align}
We seek a seperable solution where $u(x,y) = X(x)Y(y)$.
Upon substituting this form for $u$, we obtain the equality
\begin{align}
    - \frac{X'' + 2\mathrm{i}\theta_1 X + (\mathrm{i}\theta_1)^2 X}{X}
    &=
    \frac{Y'' + 2\mathrm{i}\theta_2 Y + \left((\mathrm{i}\theta_2)^2 + \omega^2\right)Y}{Y},
\end{align}
which holds at all $(x,y)\in\Omega$.
Since both sides are functions of independant variables, they must equal some constant which we label $c^2\in\mathbb{C}$.
This provides us with an ODE in each of $X$ and $Y$ to solve:
\begin{align}
    0 &= X'' + 2\mathrm{i}\theta_1 X + \left[(\mathrm{i}\theta_1)^2 + c^2 \right] X, \\
    0 &= Y'' + 2\mathrm{i}\theta_2 Y + \left[(\mathrm{i}\theta_2)^2 + \omega^2 - c^2\right]Y.
\end{align}
We can also reformulate the (Steklov) boundary condition as follows:
\begin{align}
    X(0) = -\lambda X'(0), &\quad X\left(\frac{1}{2}\right) = \lambda X'\left(\frac{1}{2}\right), \\
    Y(0) = -\lambda Y'(0), &\quad Y\left(\frac{1}{2}\right) = \lambda Y'\left(\frac{1}{2}\right).
\end{align}
Without loss of generality, we examine the system for $X$ (since $Y$ is handled similarly).
We deduce that solutions only exist when all of the following hold;
\begin{align}
    1 - \mathrm{i}\lambda\left( c - \theta_1 \right) \neq 0, &\quad
    1 + \mathrm{i}\lambda\left( c - \theta_1 \right) \neq 0, \\
    1 - \mathrm{i}\lambda\left( c + \theta_1 \right) \neq 0, &\quad
    1 + \mathrm{i}\lambda\left( c + \theta_1 \right) \neq 0, \\
    c\neq 0. &
\end{align}
Additionally we find that $\lambda$ and $c$ must be related via
\begin{align}
    \lambda &= \frac{1}{\theta_1^2 - c^2}\frac{1}{\sin\left(\frac{c}{2}\right)}
    \left[ 2c\cos\left(\frac{c}{2}\right) \pm \sqrt{c^2 - \theta_1^2\sin^2\left(\frac{c}{2}\right)} \right],
\end{align}
and that the solution $X(x)$ has the form
\begin{align}
    X(x) = \mathrm{e}^{-\mathrm{i}\theta_1 x}\left( A^+\mathrm{e}^{\mathrm{i}cx} + A^-\mathrm{e}^{-\mathrm{i}cx} \right),
\end{align}
where $A^+$ and $A^-$ solve
\begin{align}
    0 &= A^+\left[ 1 + \mathrm{i}\lambda(c-\theta_1) \right] + A^-\left[ 1 - \mathrm{i}\lambda(c+\theta_1) \right], \\
    0 &= A^+\mathrm{e}^{\frac{\mathrm{i}c}{2}}\left[ 1 - \mathrm{i}\lambda(c-\theta_1) \right] + A^-\mathrm{e}^{-\frac{\mathrm{i}c}{2}}\left[ 1 + \mathrm{i}\lambda(c+\theta_1) \right].
\end{align}

The system for $Y$ is almost identical, but with $c^2$ replaced by $\omega^2-c^2 =: \gamma^2$ and $\theta_2$ in place of $\theta_1$.
Again, we can deduce that
\begin{align}
    1 - \mathrm{i}\lambda\left( \gamma - \theta_2 \right) \neq 0, &\quad
    1 + \mathrm{i}\lambda\left( \gamma - \theta_2 \right) \neq 0, \\
    1 - \mathrm{i}\lambda\left( \gamma + \theta_2 \right) \neq 0, &\quad
    1 + \mathrm{i}\lambda\left( \gamma + \theta_2 \right) \neq 0, \\
    c^2 \neq \omega^2. &
\end{align}
Additionally we find that $\lambda$ and $\gamma$ must be related via
\begin{align}
    \lambda &= \frac{1}{\theta_2^2 - \gamma^2}\frac{1}{\sin\left(\frac{\gamma}{2}\right)}
    \left[ 2\gamma\cos\left(\frac{\gamma}{2}\right) \pm \sqrt{\gamma^2 - \theta_2^2\sin^2\left(\frac{\gamma}{2}\right)} \right],
\end{align}
and that the solution $Y(y)$ has the form
\begin{align}
    Y(y) = \mathrm{e}^{-\mathrm{i}\theta_2 y}\left( B^+\mathrm{e}^{\mathrm{i}\gamma y} + B^-\mathrm{e}^{-\mathrm{i}\gamma y} \right),
\end{align}
where $B^+$ and $B^-$ solve
\begin{align}
    0 &= B^+\left[ 1 + \mathrm{i}\lambda(\gamma-\theta_2) \right] + B^-\left[ 1 - \mathrm{i}\lambda(\gamma+\theta_2) \right], \\
    0 &= B^+\mathrm{e}^{\frac{\mathrm{i}\gamma}{2}}\left[ 1 - \mathrm{i}\lambda(\gamma-\theta_2) \right] + A^-\mathrm{e}^{-\frac{\mathrm{i}\gamma}{2}}\left[ 1 + \mathrm{i}\lambda(\gamma+\theta_2) \right].
\end{align}

As such, we deduce that the values of $\lambda$ must match for some $c$ for there to be eigenvalues.
That is, we can deduce that an eigenfunction with a seperable form exists provided that there exists some $c$ and selection $\alpha,\beta\in\{0,1\}$ such that
\begin{align}
    \left( \theta_1^2 - c^2 \right)\sin\left(\frac{c}{2}\right)\left[ 2\gamma\cos\left(\frac{\gamma}{2}\right) + \left(-1\right)^{\alpha}\sqrt{\gamma^2 - \theta_2^2\sin^2\left(\frac{\gamma}{2}\right)} \right]
    &= 
    \left( \theta_2^2 - \gamma^2 \right)\sin\left(\frac{\gamma}{2}\right)\left[ 2c\cos\left(\frac{c}{2}\right) + \left(-1\right)^{\beta}\sqrt{c^2 - \theta_1^2\sin^2\left(\frac{c}{2}\right)} \right].
\end{align}

Our goal is thus clear - we will find some values of $c$ for which we have seperable solutions, and reconstruct the eigenfunctions.
Of course, there are likely to be multiple such values of $c$ (I have no a priori intuition about how the equation we need to solve will behave for increasing $c$), so we'll do our best to find as many of them as we can.
The eigenvalue distribution is also particularly interesting in this regard.

```python
import numpy as np
from numpy import sin, cos, exp
from scipy.optimize import fsolve

import matplotlib.pyplot as plt
from matplotlib import rc
from mpl_toolkits.mplot3d import Axes3D
rc('text', usetex=True)
```

#### `cEqn(c, omega, theta, alpha, beta)`

Evaluates the expression
\begin{align}
    \left( \theta_1^2 - c^2 \right)\sin\left(\frac{c}{2}\right)\left[ 2\gamma\cos\left(\frac{\gamma}{2}\right) + \left(-1\right)^{\alpha}\sqrt{\gamma^2 - \theta_2^2\sin^2\left(\frac{\gamma}{2}\right)} \right]
    -
    \left( \theta_2^2 - \gamma^2 \right)\sin\left(\frac{\gamma}{2}\right)\left[ 2c\cos\left(\frac{c}{2}\right) + \left(-1\right)^{\beta}\sqrt{c^2 - \theta_1^2\sin^2\left(\frac{c}{2}\right)} \right], \qquad(*)
\end{align}
for the given values of $c, \omega, \theta, \alpha, \beta$.
Note that $\alpha,\beta\in\{0,1\}$.

Also worth noting is that if $c \geq \omega$, then we will get complex $\gamma$, solutions to $(*)$ might be harder to come by.

```python
def cEqn(c, omega, theta, alpha=0, beta=0):
    '''
    Evaluates expression (*) for the given values of c, omega, theta, alpha, and beta.
    INPUTS:
        c: (n,) float, array of c values to evaluate (*) at.
        omega: float, value of omega in (*).
        theta: (2,) float, value of the quasimomentum in (*).
        alpha, beta: int, either +1 or 0, values in (*).
    OUTPUTS:
        vals: (n,) float, array of values for (*) at the points in c.
    '''
    
    g2 = omega*omega - c*c
    gamma = np.sqrt(g2)
    # first term
    vals = (theta[0]*theta[0] - c*c) \
    * sin(c/2.) \
    * ( 2.*gamma*cos(gamma/2.) + ((-1.)**alpha)*np.sqrt(g2 - theta[1]*theta[1]*sin(gamma/2.)*sin(gamma/2.)) )
    # second term
    vals -= (theta[1]*theta[1] - g2) \
    * sin(gamma/2.) \
    * ( 2.*c*cos(c/2.) + ((-1.)**beta)*np.sqrt(c*c - theta[0]*theta[0]*sin(c/2.)*sin(c/2.)) )
    return vals
```

#### `Eval(c, theta, pmFactor=0, formula=0, omega=0)`

Computes the eigenvalue at the value $c$ for the given quasimomentum $\theta$.
If `formula=0` (the default), $\omega$ is not required and the formula
\begin{align}
    \lambda &= \frac{1}{\theta_1^2 - c^2}\frac{1}{\sin\left(\frac{c}{2}\right)}
    \left[ 2c\cos\left(\frac{c}{2}\right) + \left(-1\right)^{\beta} \sqrt{c^2 - \theta_1^2\sin^2\left(\frac{c}{2}\right)} \right], \qquad(l1)
\end{align}
will be used to determine $\lambda$.
If instead `formula=1`, then $\omega$ is required and the formula
\begin{align}
    \lambda &= \frac{1}{\theta_2^2 - \gamma^2}\frac{1}{\sin\left(\frac{\gamma}{2}\right)}
    \left[ 2\gamma\cos\left(\frac{\gamma}{2}\right) + \left(-1\right)^{\alpha} \sqrt{\gamma^2 - \theta_2^2\sin^2\left(\frac{\gamma}{2}\right)} \right], \qquad(l2)
\end{align}
will be used instead.
The variable `pmFactor` is used to set the value of $\alpha$ and $\beta$ in the formula being used.

Note $c$, $\theta$, and $\omega$ can be passed in as vectors, but the same formula must be used in each case.

```python
def Eval(c, theta, pmFactor=0, formula=1, omega=0):
    '''
    Evaluates the value of the Steklov eigenvalue lamba via (l1) or (l2).
    INPUTS:
        c: (n,) float, array of c values to evaluate (l1) or (l2) at.
        theta: (n,2) float, value of the quasimomentum.
        pmFactor: int, either 0 or 1, the value of alpha/beta in (l2)/(l1) respectively.
        formula: int, either 1 or 2, changing which of (l1) or (l2) to use. If 1, omega is required.
        omega: (n,) float, value of omega to evaluate (l2) at if formula is =1.
    OUTPUTS:
        lbda: (n,) float, values of the Steklov eigenvalues at the given c values.
    '''
    
    # set theta^2 values, since theta doesn't appear on it's own in the expressions
    if theta.ndim==1:
        t1 = theta[0]*theta[0]; t2 = theta[1]*theta[1]
    elif theta.ndim==2:
        t1 = theta[0,:]*theta[0,:]; t2 = theta[1,:]*theta[1,:]
    else:
        raise ValueError('theta.ndim > 2, got %d' % (theta.ndim))
    
    # compute eigenvalues
    if formula==1:
        # use (l1)
        lbda = 2.*c*cos(c/2.) + ((-1)**pmFactor)*np.sqrt(c*c - t1*sin(c/2.)*sin(c/2.))
        lbda /= (t1 - c*c) * sin(c/2.)
    elif formula==2:
        # use (l2)
        g2 = omega*omega - c*c
        gamma = np.sqrt(g2)
        lbda = 2.*gamma*cos(gamma/2.) + ((-1)**pmFactor)*np.sqrt(g2 - t2*sin(gamma/2.)*sin(gamma/2.))
        lbda /= (t2 - g2) * sin(gamma/2.)
    else:
        raise ValueError('Unknown formula requested for e-val computation, got %d' % formula)
    return lbda
```

#### `BinSearchPrep(F, xRange)`

Given a function $F$ and a range of (ordered) values `xRange` for the argument of $F$, evaluate $F$ at each point and return the index of the first of two values in xRange that lie either side of a sign change.
If there are multiple sign changes, return an index for each.
If there are no sign changes, return an error.

```python
def BinSearchPrep(F, xRange):
    '''
    Computes the value of the function F at each point in xRange, 
    returning a list of indices j for which xRange[j] and xRange[j+1] lie across a sign change in F.
    INPUTS:
        F: function, taking one (vectorised) argument x.
        xRange: (n,), arguments in order to be passed to F.
    OUTPUTS:
        inds: (m,) int, containing the m indices corresponding to sign changes in F.
    '''
    
    inds = []
    i = 0
    Fx = F(xRange)
    while i<len(Fx)-1:
        # search along the array to find sign changes
        if Fx[i]*Fx[i+1]<=0:
            # opposite signs in consecutive terms, i is an index we want
            inds.append(i)
        i += 1
    
    if not inds:
        # if inds is empty, this will execute and return the error of no sign changes found
        raise ValueError('No sign change found')
    return np.asarray(inds, dtype=int)
```

#### `ConstructSolution(c, lbda, theta, norm=True, type=0, omega=0)`

Given the value of $c$ that solves $(*)$ at $\omega, \theta$, construct the function $X(x)$ or $Y(y)$ in accordance with `type`.
If `type` is 0, then $X$ will be constructed, and a function handle to evaluate $X$ will be returned.
If `type` is 1, then $Y$ will be constructed instead.

The solution for $X$ is given by
\begin{align}
    X(x) = \mathrm{e}^{-\mathrm{i}\theta_1 x}\left( A^+\mathrm{e}^{\mathrm{i}cx} + A^-\mathrm{e}^{-\mathrm{i}cx} \right),
\end{align}
where $A^+$ and $A^-$ solve
\begin{align}
    0 &= A^+\left[ 1 + \mathrm{i}\lambda(c-\theta_1) \right] + A^-\left[ 1 - \mathrm{i}\lambda(c+\theta_1) \right], \\
    0 &= A^+\mathrm{e}^{\frac{\mathrm{i}c}{2}}\left[ 1 - \mathrm{i}\lambda(c-\theta_1) \right] + A^-\mathrm{e}^{-\frac{\mathrm{i}c}{2}}\left[ 1 + \mathrm{i}\lambda(c+\theta_1) \right].
\end{align}
In addition, we must have that the following are all satisfied for there to be _any_ solution at all:
\begin{align}
    1 - \mathrm{i}\lambda\left( c - \theta_1 \right) \neq 0, &\quad
    1 + \mathrm{i}\lambda\left( c - \theta_1 \right) \neq 0, \\
    1 - \mathrm{i}\lambda\left( c + \theta_1 \right) \neq 0, &\quad
    1 + \mathrm{i}\lambda\left( c + \theta_1 \right) \neq 0, \\
    c\neq 0. &
\end{align}
The condition $c\neq 0$ should have been filtered out prior to running this function, but it doesn't hurt to check again.
Given the above, we know that we can express $X(x)$ as
\begin{align}
    X(x) = A^+\mathrm{e}^{-\mathrm{i}\theta_1 x}\left( \mathrm{e}^{\mathrm{i}cx} - \frac{1 + \mathrm{i}\lambda(c-\theta_1)}{1 - \mathrm{i}\lambda(c+\theta_1)} \mathrm{e}^{-\mathrm{i}cx} \right),
\end{align}
which will give us our solution.
WLOG, we will set A^+ = 1, although we might be able to compute the integral actually to normalise the eigenfunction.
For the record,
\begin{align}
    \left( A^+ \right)^{-1} 
    &= \frac{\mathrm{i}}{c-\theta_1}\left( 1 - \Gamma\mathrm{e}^{\frac{\mathrm{i}(c-\theta_1)}{2}} \right)
    + \frac{\mathrm{i}}{c+\theta_1}\left( 1 - \Gamma\mathrm{e}^{-\frac{\mathrm{i}(c+\theta_1)}{2}} \right), \\
    \Gamma &= \frac{1 + \mathrm{i}\lambda(c-\theta_1)}{1 - \mathrm{i}\lambda(c+\theta_1)},
\end{align}
so we can include a normalisation method if we want to.

When $Y$ is to be constructed, the same formulae and conditions are used but with $X$ replaced by $Y$ and $c^2$ becoming $\gamma^2 = \omega^2 - c^2$.

```python
def ConstructSolution(c, lbda, theta, norm=True, type=0, omega=None):
    '''
    Constructs the solution X(x) or Y(y), depending on type.
    INPUTS:
        c: float, value of c that solves (*) at theta, omega.
        lbda: float, eigenvalue corresponding to c.
        theta: (2,) float, value of the quasimomentum.
        norm: bool, if True then eigenfunction will be normalised.
        type: int, if 0 then X is constructed, if 1 then Y is constructed. If 1, omega is required.
        omega: float, value of omega in (*). Required if type=1.
    OUTPUTS:
        F: function, the function X(x) or Y(y) depending on type.
    '''
    
    # which theta and c/gamma do we need to use?
    if type==0:
        # constructing X, uses theta[0] and c
        t = theta[0]
        C = c
    elif type==1:
        # constructing Y, uses theta[1] and gamma
        t = theta[1]
        C = np.sqrt(omega*omega - c*c)

    # check that solution is legal
    checks = np.ones((4,), dtype=complex)
    checks[0] -= 1.j*lbda*(C - t)
    checks[1] -= 1.j*lbda*(C + t)
    checks[2] += 1.j*lbda*(C - t)
    checks[3] += 1.j*lbda*(C + t)
    checkCond = np.min(np.abs(checks))
    if checkCond < 1e-8:
        # one or more of these is too close to 0, so this solution isn't OK
        raise ValueError('Solution fails to adhere to coefficient checks (A+, A-)')
    elif np.abs(C) < 1e-8:
        # the value of C itself is close to 0, so this solution isn't OK either
        raise ValueError('c is approximately 0 or omega, which are not permitted')
    
    # if we got to here, it's safe to proceed. Let's compute Gamma, and Aplus if we need it
    Gamma = checks[2] / checks[1]
    if norm:
        Aplus = 1.j*(1. - Gamma*exp(0.5j*(t-C))) / (C - t)
        Aplus += 1.j*(1. - Gamma*exp(-0.5j*(t+C))) / (C + t)
        Aplus = 1./Aplus
    else:
        Aplus = 1.
    
    # now we can return the function!
    def F(x):
        '''
        Constructed by ConstructSolution. Takes arguments 0 <= x <= 0.5.
        '''
        return Aplus * exp(-1.j*t*x) * ( exp(1.j*c*x) - Gamma * exp(-1.j*c*x) )
    
    return F
```

## Script Starts Here

```python
# close open figures in case of rerun bloat
plt.close('all')
# set some parameters
omega = 4*np.pi
theta=np.array([0.,0.])
c = np.linspace(0.,omega,num=10000,dtype=float)
# plotting things
L = 10 #number of levels in contour plot
N = 1000 #number of meshpoints for plot

for alpha in [0,1]:
    for beta in [0,1]:
        # setup function handle
        F = lambda x: cEqn(x, omega, theta, alpha=alpha, beta=beta)
        # narrow down search range for initial guess
        try:
            cInds = BinSearchPrep(F, c)
        except ValueError:
            raise ValueError('Found no indices to use for starting guess in binary search!')
        # begin printouts
        print('-------------- \n alpha = %d, beta = %d' % (alpha, beta))
        print(' Found %d sign changes ' % len(cInds), cInds)
        # loop over all sign changes to find solution values
        for i, ii in enumerate(cInds):
            # if c=0 and c=omega have been caught, we need to exclude these
            if (c[ii]>0 and c[ii+1]<omega):
                cSolve = fsolve(F, (c[ii]+c[ii+1])/2.)
                lbdaA = Eval(cSolve, theta, pmFactor=alpha, formula=2, omega=omega)
                lbdaB = Eval(cSolve, theta, pmFactor=beta, formula=1)
                print(' [%d]  Found c = %.5f, \n lambda (l1): %.5f, lambda (l2): %.5f' % (i, cSolve, lbdaB, lbdaA))
                if np.abs(lbdaA-lbdaB)>=1e-8:
                    print('E-VAL MISMATCH')
                else:
                    # proceed with solution construction
                    X = ConstructSolution(cSolve, lbdaA, theta, norm=True, type=0)
                    Y = ConstructSolution(cSolve, lbdaB, theta, norm=True, type=1, omega=omega)
                    u = lambda x,y: X(x)*Y(y)
                    
                    # plot data setup
                    xRange = yRange = np.linspace(0.,0.5,num=N)
                    xx, yy = np.meshgrid(xRange, yRange)
                    uData = u(xx, yy)
                    
                    # figure setup
                    rFig, rAx = plt.subplots()
                    iFig, iAx = plt.subplots()
                    # axis setup
                    for a in [rAx, iAx]:
                        a.set_aspect('equal')
                        a.set_xlabel(r'$x_1$')
                        a.set_ylabel(r'$x_2$')
                    rAx.set_title(r'$\Re (u), \ \lambda = %.5f $' % lbdaA)
                    iAx.set_title(r'$\Im (u), \ \lambda = %.5f $' % lbdaA)
                    # make contour plots
                    rCon = rAx.contourf(xRange, yRange, np.real(uData), levels=L)
                    iCon = iAx.contourf(xRange, yRange, np.imag(uData), levels=L)
                    # make colourbars
                    rFig.colorbar(rCon)
                    iFig.colorbar(iCon)
                    rFig.show()
                    iFig.show()
            else:
                print(' [%d]  c=0 or c=omega not allowed' % i)
```
